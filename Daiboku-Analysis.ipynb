### Data Wrangling

##### First we find all missing data

```python
df.isna().sum()
```

We see that Payment_Type and Category have a lot of missing data. Let us investigate.

```python
df.loc[df['Category'].isna()]
```

We see that where there is missing data in the Category column, there is a payment type, which implies that there is an Order_Total. Thus, it would not make sense to drop these rows with missing values.

```python
df.loc[df['Category'].isna(), 'Payment_Type'].value_counts()
```

Whenever payment type  is NaN, this means that the customer did not pay out. Thus we cannot count this as a sale. We can verify this information seeing 'Exchange_Void_or_Return' for the Order_Totals that have a value greater than 0. If payment type is NaN and Order_Total is 0.00, there is no need to Void or Return. 

```python
df.loc[(df['Payment_Type'].isna()) & (df['Order_Total'] != 0.00)]
```

```python
df.loc[(df['Payment_Type'].isna()) & (df['Order_Total'] == 0.00)]
```

We should also drop rows if Exchange_Void_or_Return is 1 because it does not count as a sale. It also would not make sense to replace null values with mode/mean. 

```python
df['Exchange_Void_or_Return'].replace({1: None}, inplace=True)
df.dropna(subset=['Payment_Type', 'Exchange_Void_or_Return'], inplace=True)
```

We must now change Order_Date to datetime format.

```python
pd.to_datetime(df['Order_Date'])
```

There is no reason to remove duplicates in this dataset in any of the columns, but below is a count of all duplicates in each column.
